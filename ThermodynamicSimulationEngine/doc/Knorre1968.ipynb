```python
fracoccup = np.genfromtxt('/tmp/Knorre1968fracoccup.csv', delimiter=',')
```

The integrator asks for fractional occupancy in the past (because of lags). So the time goes there and back all the time. Need to sort it in order

```python
values = fracoccup[:,1]
times = fracoccup[:,0]
d = zip(times, values)

s = [i[1] for i in sorted(d)]
```

## The fractional occupancy smoothly goes from being quite low to being still quite low

```python
plot(sorted(times),s)
show()
```

```python
plot(sorted(times),s)
xlim([-0.1,0.1])
```
