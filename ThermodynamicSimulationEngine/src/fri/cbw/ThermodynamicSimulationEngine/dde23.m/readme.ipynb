#This is a brief description of all that is related to ode23

DDE23 function is a part of Matlab, but an old version of it is available at the webpage of its authors [Soving DDEs with MATLAB](ttp://www.radford.edu/~thompson/webddes/) . They have a **good tutorial** there that explains how to work with it from the user point of view. The tutorial in PDF is in the same directory as this file. It may seem too basic at times but reading it at least up to the second example is a good idea to learn how the solver works and it gives simple examples of what DDE's are.

The only limitation of the function is that it can only work with constant (predetermined) delays. There is a version without this restriction, but for our use constant delays suffice.

The Java implementation in class ODE23 has number of additional restrictions since I put in only what was necessary for this work. The restrictions are listed in the JavaDoc, but essentially it means there are no configurable options.

Usage of the class can be inferred by looking at **Example 1** in prevously mentioned tutorial and its implementation as the first Unit Test.

## Example 1

```python
%%octave
function v = exam1f(t,y,Z)
    ylag1 = Z(:,1);
    ylag2 = Z(:,2);
    v = zeros(3,1);
    v(1)=ylag1(1);
    v(2)=ylag1(1) +y(2);
    v(3)=ylag2(2);
end
function v = exam1h(t)
    v = ones(3,1);
end
```

```python
%%octave
sol1 = dde23(@exam1f,[1, 0.2],ones(3,1),[0, 5]);
plot(sol1.x, sol1.y)
```

## Blowfly

```python
%%octave
function v = blowflyf(t,y,Z)
    c = [12.0, 0.25, 10.0, 300.0, 100.0];
    alag = 0.0;
    if (t > c(1))
        alag = Z(1);
    end
    v = [c(3)*alag*exp(-alag/c(4))-c(2)*y(1)];
end
```

```python
%%octave
sol2 = dde23(@blowflyf,[12],[300],[0, 300]);
plot(sol2.x, sol2.y);
```

## Lotka-Volterra

```python
%%octave
function v = LotkaVolterraf(t,y,Z)
    alpha = 1;
    beta = 0.2;
    gamma = 0.5;
    delta = 0.04;
    v = [y(1) * (alpha - beta*y(2));-y(2) * (gamma - delta*y(1))];
end
```

```python
%%octave
sol3 = dde23(@LotkaVolterraf,[],[5,2],[0, 50]);
plot(sol3.x, sol3.y);
```

## ddesolver for Python

```python
from numpy import array, exp
from PyDDE import pydde
dde_eg = pydde.dde()

def ddegrad(s, c, t):
    alag = 0.0
    if (t>c[0]):
        alag = pydde.pastvalue(0,t-c[0],0)
    return array( [ c[2]*alag*exp(-alag/c[3])-c[1]*s[0] ] )

def ddesthist(g, s, c, t):
    return (s, g)

ddecons = array([12.0,0.25,10.0,300.0,100.0])
ddeist = array([ddecons[4]])
ddestsc = array([0])

dde_eg.initproblem(no_vars=1, no_cons=5, nlag=1, nsw=0,
                   t0=0.0, t1=300.0,
                   initstate=ddeist, c=ddecons,
                   otimes=numpy.arange(0.0, 300.0, 1.0),
                   grad=ddegrad, storehistory=ddesthist)

dde_eg.initsolver(tol=0.000005, hbsize=1000,
dt=1.0, statescale=ddestsc)

sol1 = dde_eg.solve()
```

```python
%matplotlib inline
import matplotlib
plt = matplotlib.pyplot
```

```python
plt.plot(sol1)
plt.show()
```

```python
sol2 = dde_eg.dde(y=ddeist, times=numpy.arange(0.0, 300.0, 1.0),
func=ddegrad, parms=ddecons,
tol=0.000005, dt=1.0, hbsize=1000, nlag=1, ssc=ddestsc)
```
